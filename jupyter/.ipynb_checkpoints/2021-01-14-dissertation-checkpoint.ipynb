```python
def blah(df):
    a = pd.DataFrame(df.iloc[:, 2:].values * df.iloc[:, 1:2].values, columns=df.columns[2:])
    return a.mean(axis=0).to_frame()
```

```python
rv_by_variation = tmp.groupby('file').apply(blah).unstack(-1)[0]
```

```python
def create_random_normal(mean, std, n):
    return np.random.normal(mean, std, int(n))
```

```python
stats.columns
```

```python
# a.reset_index(-1).drop('level_2', axis=1)
out = []
for metric in ['rv', 'r10_iso', 'rq', 'ra', 'rz']:
    a = stats.groupby(['file', 'Variation'])['ra'].agg(['mean', 'std', 'count'])
    res = a.groupby(level=-1).agg({'mean': 'mean', 'std': 'sum', 'count': 'sum'})
    out.append(
        pd.DataFrame([
        [variation,
         nb.st.ttest_ind(
             create_random_normal(*res.loc[variation]),
             create_random_normal(*res.loc['Original'])
         ).pvalue] 
        for variation in res.index[:-1].values
    ], columns=['Variation', 'p-value']).assign(metric=metric)
    )
```

```python
tmp = pd.concat(out).set_index(['Variation', 'metric', ]).unstack(-1)['p-value'].round(2)
```
